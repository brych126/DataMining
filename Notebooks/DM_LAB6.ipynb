```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/brych126/DataMining/main/movies_cleaned.csv'
dataset = pd.read_csv(url, encoding='latin-1')
dataset.head() 
```

###2.Оберіть метрику яку ви використаєте для ваших даних і задачі  та аргументуйте чому обрали її 


**Середня абсолютна помилка: MAE** - це абсолютна різниця між цільовим значенням та значенням, передбаченим моделлю. **MAE** є більш надійним для **outliers** і не настільки критичним до помилок як **MSE** та **RMSE**. **MAE** - це лінійна оцінка, що означає, що всі індивідуальні відмінності зважуються однаково. Він не підходить для додатків, де ви хочете приділити більше уваги **outliers**.

###3.Використайте вашу обрану метрику на натренованій моделі на тестових даних. Наведіть результати метрики для моделей з 4 і 5 лабораторної
 

```
#Regression
from sklearn.model_selection import train_test_split
x1 = dataset[['budget', 'runtime', 'gross']]
y1 = dataset['score']
x_train, x_test, y_train, y_test = train_test_split(x1, y1, test_size=0.2, random_state=0)
print('Data was split to train and test sets\nTrain set size:',len(x_train),'\nTest set size:', len(x_test))
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
from sklearn import metrics
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```
x_train2, x_test2, y_train2, y_test2 = train_test_split(x1, y1, test_size=0.2, random_state=0)
print('Data was split to train and test sets\nTrain set size:',len(x_train),'\nTest set size:', len(x_test))
from sklearn import linear_model
regressor = linear_model.Lasso(alpha=0.1)
regressor.fit(x_train2, y_train2)
y_pred2 = regressor.predict(x_test2)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test2, y_pred2))
print('Mean Squared Error:', metrics.mean_squared_error(y_test2, y_pred2))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test2, y_pred2)))
```
